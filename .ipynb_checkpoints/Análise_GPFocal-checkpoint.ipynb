## Dividir e Filtrar Texto

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
from sklearn.cluster import KMeans
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
```

```python
sentencas = [sentenca for paragrafo in sentencas for sentenca in paragrafo]
```

```python
nltk.download('stopwords')
```

```python
stop_words_pt = set(stopwords.words('portuguese'))
```

```python
vectorizer = TfidfVectorizer(stop_words=list(stop_words_pt))
```

```python
X = vectorizer.fit_transform(sentencas)
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
kmeans.fit(X)
```

```python
clusters = kmeans.labels_
```

```python
#for i in range(max(clusters) + 1):
  #  print(f"Cluster {i + 1}:")
 #   for j, sentenca in enumerate(sentencas):
#        if clusters[j] == i:
 #           print(f"  - {sentenca}")
  #  print("\n")
```

```python
cluster_desejado = 3
```

```python
sentencas_cluster1 = []
```

```python
for j, sentenca in enumerate(sentencas):
  if clusters[j] == cluster_desejado:
      sentencas_cluster1.append(sentenca)
```

## Identificação de Tópicos

```python
from gensim import corpora, models
```

```python
from nltk.tokenize import word_tokenize
```

```python
from nltk.corpus import stopwords
```

```python
nltk.download('punkt')
```

```python
palavras_a_ignorar = ["participante", "mediador", "6", "5", "4", "3", "2", "1"]
```

```python
stop_words_pt = set(stopwords.words('portuguese') + palavras_a_ignorar)
```

```python
sentencas_processadas = [
  [token.lower() for token in word_tokenize(sentenca) if token.isalnum() and token.lower() not in stop_words_pt]
  for sentenca in sentencas_cluster1
]
```

```python
dicionario = corpora.Dictionary(sentencas_processadas)
```

```python
corpus = [dicionario.doc2bow(sentenca) for sentenca in sentencas_processadas]
```

```python
lda_model = models.LdaModel(corpus, num_topics=5, id2word=dicionario, passes=100, random_state=42)
```

```python
print("Tópicos identificados:")
for i, topic in lda_model.print_topics(num_words=7):
  print(f"Tópico {i + 1}: {topic}")
```

```python
G = nx.DiGraph()

G.add_nodes_from([
    ("Infoprodutos", {"cor": "aquamarine"}),
    ("Escolha", {"cor": "pink"}),
    ("Experiência", {"cor": "pink"}),
    ("Formato", {"cor": "pink"}),
    ("Aprendizagem", {"cor": "pink"}),
    ("Interação", {"cor": "skyblue"}),
    ("Desafio", {"cor": "lightblue"}),
    ("Opção", {"cor": "lightblue"}),
    ("Prazo", {"cor": "lightblue"}),
    ("Projeto", {"cor": "lightblue"}),
    ("Produção", {"cor": "lightblue"}),
    ("Influência", {"cor": "lightblue"}),
    ("Acomodação", {"cor": "lightblue"}),
    ("Específico", {"cor": "lightblue"}),
    ("Avançado", {"cor": "lightblue"}),
    ("Conteúdo", {"cor": "lightblue"}),
    ("Espaço", {"cor": "lightblue"}),
    ("Consumo", {"cor": "lightblue"}),
])

G.add_edges_from([
    ("Infoprodutos", "Escolha"),
    ("Infoprodutos", "Experiência"),
    ("Infoprodutos", "Formato"),
    ("Infoprodutos", "Aprendizagem"),
    ("Escolha", "Desafio"),
    ("Escolha", "Opção"),
    ("Escolha", "Prazo"),
    ("Experiência", "Projeto"),
    ("Experiência", "Produção"),
    ("Experiência", "Influência"),
    ("Experiência", "Acomodação"),
    ("Formato", "Específico"),
    ("Formato", "Avançado"),
    ("Formato", "Conteúdo"),
    ("Aprendizagem", "Interação"),
    ("Interação", "Espaço"),
    ("Interação", "Consumo"),
])

pos = {
    "Infoprodutos": (0, 0),
    "Escolha": (-1, 1),
    "Desafio": (-1.5, 2),
    "Opção": (-0.5, -1),
    "Prazo": (-1.5, 0),
    "Experiência": (1, 1),
    "Projeto": (0.5, 2),
    "Produção": (2, 2),
    "Influência": (2.5, 1.5),
    "Acomodação": (2, 0.5),
    "Formato": (1, -1),
    "Específico": (0, -2),
    "Avançado": (1, -2),
    "Conteúdo": (2, -2),
    "Aprendizagem": (1.5, 0),
    "Interação": (3, 0),
    "Espaço": (2.5, 1),
    "Consumo": (2.5, -1),
}

node_color = [G.nodes[n]["cor"] for n in G.nodes]

plt.figure(figsize=(8, 6))
nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=700,
    node_color=node_color,
    font_size=10,
    font_color="black",
    font_weight="bold",
    arrowsize=20,
    edge_color="gray",
    linewidths=5,
    width=0.5,
)

plt.title("Mapa Mental: Infoprodutos", fontsize='14', fontweight='bold')
plt.show()
```

```python
nltk.download('vader_lexicon')
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
topic_keywords = lda_model.print_topics(num_words=100)
```

```python
topic1_keywords = topic_keywords[0][1]
topic2_keywords = topic_keywords[1][1]
topic3_keywords = topic_keywords[2][1]
topic4_keywords = topic_keywords[3][1]
topic5_keywords = topic_keywords[4][1]
```

```python
topic1_keywords_list = [word.split("*")[1] for word in topic1_keywords.split(" + ")]
```

```python
from googletrans import Translator

# Função para traduzir uma lista de palavras para inglês
def translate_to_english(word_list):
    translator = Translator()
    english_translations = [translator.translate(word, dest='en').text for word in word_list]
    return english_translations


# Traduzir para inglês
topic1_keywords_english = translate_to_english(topic1_keywords_list)

# Exibir a lista de palavras traduzidas
print(topic1_keywords_english)
```

```python
topic1_keywords_english_str = " ".join(topic1_keywords_english)
```

```python
print(topic1_keywords_english_str)
```

```python
sentiment_score = sia.polarity_scores(topic1_keywords_english_str)
```

```python
print(sentiment_score)
```
